# Gamma, Exponential, & Erlang Distributions


In a previous notebook I described how to simulate data that follows a gamma distribution and how to investigate the properties of that distribution, using the Probability Density Function and the Cumulative Distribution Function. The gamma distribution is a continuous distribution function that is used to model the time between independent events that occur at a given rate. The three main parameters for the gamma distribution are the threshold (loc), the shape, and the scale. These specify the starting value for the distribution, the number of events being modeled, and the rate at which events occur respectively. 

In this notebook I will develop our understanding of the gamma distribution by visualising the gamma distribution to show how changing these parameters alters the appearance of the distribution. 

I will also discuss the relationship between the gamma distribution and the two other distributions, the exponential distribution and the erlang distribution. These latter distributions being just special cases of the gamma distribution. 

```python
# Importing key software libraries. 

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

#### Adjusting the loc (threshold) parameter

```python
# Next I will create an array of data. This will be 100 data points on a scale between -15 and 40.
# The variable/ object x is the array of datapoints. 
# I will then use this to define three gamma distributions. In this case I will set the shape (a) and scale parameters 
# both equal to three and set the loc (threshold) parameter to three different values: -10, 0, and 10. 

x = np.linspace(-15, 40, 100)

loc_1 = stats.gamma.pdf(x, a = 3, loc = -10, scale = 3)
loc_2 = stats.gamma.pdf(x, a = 3, loc = 0, scale = 3)
loc_3 = stats.gamma.pdf(x, a = 3, loc = 10, scale = 3)

# Plotting the distribution of each line. 

plt.plot(x, loc_1, label = 'loc = -10, shape = 3, scale = 3')
plt.plot(x, loc_2, label = 'loc = 0, shape = 3, scale = 3')
plt.plot(x, loc_3, label = 'loc = 10, shape = 3, scale = 3')

plt.legend()
plt.show()
```

The figure above highlights how changing the loc (threshold) for the gamma distribution adjusts the starting value and shifts the distribution to the left or right on the x-axis. For example, we can see the distribution for the blue line starts at -10 and is the most left shifted of the three. The orange distribution shows probabilities starting at 0, and we can see the green line starts at 10, so all values lower than this have a 0 probability, hence the flat line. The green line is clearly the most right shifted distribution of the three. 

Next, we will investigate what happens when you adjust the shape parameter while keeping the other parameters constant. 

#### Adjusting the shape (a) parameter

```python
# Here I will plot three gamma distributions. I will set the loc to 0, the scale to 3, and adjust the shape (a) parameter. 
# The shape captures the number of events you are modeling. For example, the elapsed time for x number of events. 
# In the below plot I will set the shape to the values 1, 3, and 5. 
# The shape is denoted by a in the below code. 

x = np.linspace(0, 40, 100)

shap_1 = stats.gamma.pdf(x, a = 1, loc = 0, scale = 3)
shap_2 = stats.gamma.pdf(x, a = 3, loc = 0, scale = 3)
shap_3 = stats.gamma.pdf(x, a = 5, loc = 0, scale = 3)

# Plotting the distribution of each line. 

plt.plot(x, shap_1, label = 'loc = 0, shape = 1, scale = 3')
plt.plot(x, shap_2, label = 'loc = 0, shape = 3, scale = 3')
plt.plot(x, shap_3, label = 'loc = 0, shape = 5, scale = 3')

plt.legend()
plt.show()
```

The above probability density function plot for the three gamma distribution shows how the appearance of the distribution changes as the value of the shape (a) parameter changes. It can be seen that increasing the value of the shape parameter decreases the height of the distribution and widens it. So, the distributions become less positively skewed as the shape parameter increases. When the value of the shape parameter is very high the gamma distribution will start to take on the appearance and approximate a normal (Gaussian) distribution, as shown below. 

```python
# Plotting a gamma distribution with a high shape value (a = 50) on a scale between 0 and 300.

x = np.linspace(0, 300, 1000)

shap_4 = stats.gamma.pdf(x, a = 50, loc = 0, scale = 3)

plt.plot(x, shap_4, label = 'loc = 0, shape = 50, scale = 3')

plt.legend()
plt.show()
```

Above, we can see that with a high shape value (a = 50) the gamma distribution starts to take on a normal bell-shaped appearance. The scale used above is between 0 and 300 time units.  

#### Adjusting the scale parameter

```python
# To investigate the effects of adjusting the scale parameter I will again plot three gamma distributions. 
# I will set the loc to 0, the shape to 4, and adjust the scale parameter. 
# The scale captures the rate or mean amount of time between events. As a rate it is often specified as 1/ lambda.  
# In the below plot I will set the scale to the values 1, 3, and 5. 
# The x-axis scale for the data is set between 0 and 60 

x = np.linspace(0, 60, 100)

scal_1 = stats.gamma.pdf(x, a = 4, loc = 0, scale = 1)
scal_2 = stats.gamma.pdf(x, a = 4, loc = 0, scale = 3)
scal_3 = stats.gamma.pdf(x, a = 4, loc = 0, scale = 5)

# Plotting the distribution of each line. 

plt.plot(x, scal_1, label = 'loc = 0, shape = 4, scale = 1')
plt.plot(x, scal_2, label = 'loc = 0, shape = 4, scale = 3')
plt.plot(x, scal_3, label = 'loc = 0, shape = 4, scale = 5')

plt.legend()
plt.show()
```

Similar to adjusting the shape parameter, higher values of the scale parameter reduce the height of the distribution and widen the distribution. This makes sense when we consider that the scale captures the mean time between events. When the time between events is longer (higher scale values) the probability of higher values for elapsed time increases. 

## Exponential Distribution

The exponential distribution is a continuous probabaility distribution that we use to model the time elapsed (waiting time) until one event occurs. As such, the exponential distribution is just a special case of the gamma distribution. A gamma distribution with a shape (a) value of 1 is equivalent to an exponential distribution. So, the exponential distribution has one less parameter than the gamma distribution as the shape (a) is set to one by default. 

Using the scipy software library we can generate data that follows an exponential distribution using the expon method, where we specify the scale (rate) parameter and the size of the sample of interest. We can also use the other methods available through scipy to investigate probabilities for a continuous variable that follows an exponential distribution using the pdf and cdf. 

```python
# To simulate/ generate data that follows an exponential distribution using the scipy expon method and random variates (rvs).
# Here I am specifying a rate of 40 events per unit of time and a sample size of 10,000. 

expon_var = stats.expon.rvs(scale = 40, size = 10000)
```

```python
# We can create a histogram to visualise what the distribution of our data looks like. 

plt.hist(expon_var, density = True, edgecolor = 'black')

plt.show()
```

Although the exponential distribution is a continuous distribution, we can obtain a histogram that gives us densities based on the rate. We can likewise obtain the Probability Density Function (pdf) for our data. 

```python
# Plotting the pdf for exponential data. 
x = np.linspace(0, 400, 10000)
ex = stats.expon.pdf(x, loc = 0, scale = 40)

plt.plot(x, ex, label = 'Exponential: Rate = 40')

plt.legend()
plt.show()
```

```python
# We can also use the pdf method to obtain probabilities for given values on the x-axis. 
# For example, the probability of waiting 5 time units (e.g. minutes) before an event occurs
# with a rate of occurrence (scale) of 40. 

stats.expon.pdf(5, loc = 0, scale = 40)
```

We can see there is about a 2% probability that we would see an event in exactly 5 minutes with a rate of occurrence of 40.

```python
# Similarly we can use the cdf method to obtain probabilities that we would have to wait a given time or less before 
# observing an event when the rate is 40. In this case the probability we would see an event occur in 50 minutes or less. 
stats.expon.cdf(50, loc = 0, scale = 40)
```

So, we can see that when the scale (the mean time between events) is 40, there is a 71% probability we would observe an event within 50 minutes or less. 

### Erlang Distribution

The erlang distribution is another distribution that is also a special case of the gamma distribution. Recall that the shape (a) of the gamma distribution models the number of events you are modeling. For example, the elapsed time until 3 accidents have occurred. Whereas the exponential distribution is the gamma distribution with a shape value of 1, the erlang distribution is equivalent to the gamma distribution when the shape value is set to an integer. The shape parameter for a gamma distribution must be a positive value but need not be an integer. When it is an integer then the gamma and the erlang are identical. 

As with other distributions, we can simulate data that follows an erlang distribution using the erlang method and invoking the rvs method in scipy. Similarly, we can also investigate probabilities for this distributin using the pdf and cdf. 

```python
# Simulating data that follows an erlang distribution. Specifying the shape (a) = 3. This should be an integer
# value. If you enter an non-integer value this will not be inforced but you will get a warning message. 

erlang_var = stats.erlang.rvs(a = 3, loc = 0, scale = 1, size = 10000)
```

```python
# We can create a histogram to visualise what the distribution of our data looks like. 

plt.hist(erlang_var, density = True, edgecolor = 'black')

plt.show()
```

```python
# Plotting the pdf for erlang distributed data with the parameters used above. 
data = np.linspace(0, 40, 10000)
erl = stats.erlang.pdf(data, a = 3, loc = 0, scale = 1)

plt.plot(data, erl, label = 'Erlang with shape a = 3')

plt.legend()
plt.show()
```

```python
# We can also use the pdf method to obtain probabilities for given values on the x-axis. 
# For example, the probability of waiting exactly 5 time units (e.g. minutes) before our a = 3 events occur
# with a rate of occurrence (scale) of 1. 

stats.erlang.pdf(5, a = 3, loc = 0, scale = 1)
```

There is an 8% probability that in 5 minutes (time units) we would see our 3 events occur (shape = 3) when the rate is 1 event per time unit. 

```python
# Similarly we can use the cdf method to obtain probabilities that we would have to wait a given time or less before 
# observing  the 3 events specified by our shape when the rate is 1 per time unit. 
# In this case the probability we would see 3 events occur in 6 time units (minutes) or less. 

stats.erlang.cdf(6, a = 3, loc = 0, scale = 1)
```

The cumulative probability that we will see our modeled number of events (a = 3) in 6 time units or less is about 94%. 

### Summary:

In this notebook I have discussed the properties of the gamma distribution and how changing the parameters influences the appearance of the distribution. I have also described the exponential and erlang distributions as special cases of a gamma distribution. 

- Changing the loc (threshold) of a gamma distribution shifts the starting vlaue to the left or right on the x-axis. 
- Increasing the shape parameter (a) causes elapsed time to increase, and makes the distribution wider and flatter, this is because we are modeling more events and it will take longer before those events are observed when they occur at a fixed rate. 
- Increasing the scale parameter (rate) also increases the width of the distribution and redcues the height as the mean time between events is increasing. 
- The exponential distribution is the gamma distribution with the shape value (a) set to a default of 1. E.g. the elapsed time for 1 event to occur. 
- The erlang distribution is the gamma distribution when the shape value (a) is set to integer values.
