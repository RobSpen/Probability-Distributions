# Beta Distribution


The beta distribution is a continous probability distribution. The standard beta distribution uses the interval [0, 1] so data values used as input for a beta distribution take on values between 0 and 1. The beta distribution is used to model random variables of finite length. It is suitable to model the behaviour of percentages and proportions. As such it is often used to model the probability of probabilities. It can be utilised once we have a Bernoulli distribution, which would give us a distribution of successes and failures, that can be interpreted as the probability of having however many successes we observed out of the total number of trials. Note that the Bernoulli distribution could only take on discrete values of success or failure (counts of success/failure). However, the beta distribution (being continuous) can take on any value between 0 and 1. So, if we wanted to work out the probability of observing a given success outcome (e.g. rain occurring) when we know a success probability for that event we would use the beta distribution. For example, what is the probability of us observing rain today, if it rained on 70 days out of 100 (it rains with p = 0.7 (70% chance)).

We can use the beta distribution to calculate the probability of an event when we have some information about the volume/ proportions of successes and failures. The primary use of the beta distribution is in Bayesian statistics where it is used as a conjugate prior for the binomial and Bernoulli distributions. A conjugate prior is observed when the prior and posterior distributions come from the same family of probability distributions.

The beta distribution has shape and scale parameters like a number of other distributions. However, the beta distribution has two shape parameters $ {\alpha} $ and $ {\beta} $. Both of these parameters must be greater than 0. The shape parameters $ {\alpha} $ and $ {\beta} $ are calculated as:

$ {\alpha} $ = k + 1
$ {\beta} $ = n - k - 1,

where: 

- k = the number of successes 
- n = the number of trials

The $ {\alpha} $ and $ {\beta} $ parameters control the shape of the distribution. Adjusting the values of $ {\alpha} $ and $ {\beta} $ will change the shape of the beta distribution's Probability Density Function (pdf) and Cumulative Distribution Function (cdf). In addition, the formula for the beta distribution commonly contains the parameters a and b, which are used to denote the lower and upper bounds of the finite interval [0, 1]. 



In this notebook, I will demonstrate how to simulate data that follows a beta distribution. I will plot the pdf of the beta distribution using different values for the shape parameters ${\alpha}$ and ${\beta}$ to visualise how changing these parameters changes the appearance of the distribution. I will also use the pdf and cdf to obtain probability for specific input values. 

```python
# Importing key software libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
# Generating data using the scipy stats beta method and invoking the random variates (rvs) method. 
# For the beta method, arguments for alpha and beta are needed. We can also specify the loc and
# scale parameters. By default they will be set to loc = 0 (starting value) and scale = 1 (end value).
# We can also use the random_state argument to set a random seed. 
# I have set alpha and beta to a = 31 and b = 21 respectively. 
# The generated data consists of 1000 datapoints. 

beta_var = stats.beta.rvs(a = 31, b = 21, loc = 0, scale = 1, size = 1000, random_state = 1)
```

```python
# Plotting the simulated data. Here I have used seaborn's displot to obtain a histogram. 
# Note that the values on the x-axis, the values for the distribution, all fall between 0 and 1. 

sns.displot(beta_var)
```

```python
sns.displot(beta_var, kind = 'kde')
```

The above distribution, given our ${\alpha}$ and ${\beta}$ values looks to be fairly normally distributed in appearance. We can obtain probabilities of given observations using the pdf and cdf for a beta distribution with these shape parameters. 

```python
# Using the pdf we can find out the probability of observing a value of 0.66. Here, x = 0.66.
# Using the same alpha (a = 31) and beta (b = 21) parameters as above. 

stats.beta.pdf(x = 0.66, a =31, b = 21, loc = 0, scale = 1)
```

The above result indicates that there is about a 3.95% (p = 0.395) probability of observing a value of 0.66. 
Below, we can similarly see there is a 0.1% probability of observing the value 0.4. This makes sense when inspecting our histogram as we can see that this value is in the left tail of the distribution and will have a much lower probability. 

```python
stats.beta.pdf(x = 0.4, a =31, b = 21, loc = 0, scale = 1)
```

```python
# The cdf can also be used to tell us the probability of observing a given value or lower. 
# The cdf method takes the same arguments as the pdf. 

stats.beta.cdf(x = 0.66, a =31, b = 21, loc = 0, scale = 1)
```

Above, we see that there is about an 82.5% chance of observing a value of 0.66 or less in our beta distribution. 
Below, there is a less than 1% chance that an observation will have a value of 0.4 or less. 

```python
# Similarly with a lower values observation 0.4
stats.beta.cdf(x = 0.4, a =31, b = 21, loc = 0, scale = 1)
```

### Visualising how the beta distribution changes shape

The below plots demonstrate how the beta distribution changes shape when the values of ${\alpha}$ and ${\beta}$ change. 

```python
# Plotting the beta using matplotplib, setting the shape parameters alpha and beta both equal to 0.5. 

alpha = 0.5
beta = 0.5

x = np.arange (0, 1, 0.01)
y = stats.beta.pdf(x, alpha, beta)
plt.figure(figsize=(11,6))
plt.plot(x, y, linewidth=3)
plt.xlabel('x', fontsize=20)
plt.ylabel('PDF', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```
